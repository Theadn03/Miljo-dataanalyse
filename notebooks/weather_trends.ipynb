**Visualisering av værdata**

Vår visualisering av værdata er basert på historisk værdata mellom 01.01.2023 til 01.05.2025 fra 3 ulike byer - Steinkjer, Molde & Ålesund. 

```python
"""
Comprehensive analysis and visualization script for cleaned weather data.

This script loads the dataset, imports all relevant modules from the src directory,
and prepares the data for statistical analysis, regression modeling, and plotting.
"""

import os
import sys
sys.path.append(os.path.abspath("../src"))

import pandas as pd
from visualize_data import (
    plot_temperature_trend,
    plot_environmental_factors,
    plot_precipitation,
    plot_wind_speed
)

# Load dataset exported from main pipeline
file_path = '../data/weekly_weather_data.csv'
df = pd.read_csv(file_path)

# Ensure 'Time' column is parsed as datetime
df['Time'] = pd.to_datetime(df['Time'])

print("Data loaded")
```

```python
"""
Generate line plots for key environmental variables over time.
"""

# Plot temperature trend over time for each city
plot_temperature_trend(df)

# Plot relative humidity trend over time for each city
plot_environmental_factors(df)

# Plot precipitation amounts over time for each city
plot_precipitation(df)

# Plot wind speed over time for each city
plot_wind_speed(df)
```

**Temperaturutvikling**     
Den første grafen viser utviklingen i temperatur kl 12:00 hver mandag i hele måleperioden. Vi ser tydelige sesongvariasjoner: temperaturene stiger om våren og sommeren og synker om høsten og vinteren. Steinkjer har generelt de laveste temperaturene, spesielt vinterstid, noe som er forventet siden det ligger lengre nord og mer innlands. Ålesund viser de høyeste temperaturene, særlig i vintersesongen, noe som trolig skyldes den milde kystklimaeffekten. Molde ligger midt imellom. Det er tydelig at temperaturvariasjonene følger et naturlig mønster og gir god støtte for videre analyse.

**Luftfuktighet**       
Grafen over relativ luftfuktighet viser at alle tre byer i stor grad ligger høyt, ofte mellom 70–100 %. Steinkjer har en tendens til litt lavere verdier i perioder, noe som kan forklares av kaldere og tørrere innlandsvær. Ålesund, med sin beliggenhet ved kysten, har generelt høy luftfuktighet med mindre variasjon. Dette er i tråd med forventningene for et fuktig og mildt kystklima. Molde viser litt større variasjon i luftfuktighet sammenlignet med Ålesund, men følger et lignende mønster.

**Nedbør**          
I nedbørsgrafen ser vi at Ålesund skiller seg klart ut med hyppigere og kraftigere nedbørstopper. Dette er helt i tråd med kjente værforhold på Vestlandet, hvor fjell og hav skaper mye nedbør. Molde har også en del nedbør, men ikke like intensivt eller hyppig som Ålesund. Steinkjer har klart færrest episoder med høy nedbør, og viser et tørrere mønster generelt. Her blir det tydelig hvordan geografisk plassering påvirker nedbørsmengder kraftig.

**Vindhastighet**           
Den siste grafen viser utviklingen i vindhastighet, men kun for Steinkjer. Ved innhenting av data så er det kun registert målinger ved målestasjonen til Steinkjer i perioden januar til juli 2023. Dette er en kort periode så man har ikke mulighetentil å sammenligne resultatet med andre år, noe som begrenser muligheten for å analysere funnet. Vi ser at vindstyrken i Steinkjer oftest ligger mellom 1 og 3 m/s, med enkelte topper på 6–8 m/s i perioden. Det er tydelig at vind er mest intens i vintermånedene og roligere gjennom sommeren.
